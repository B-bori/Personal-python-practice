```
from sympy import *
import numpy as np 
from io import StringIO

a = Symbol('a')

M = np.array([[2,0], [0,2]])

''' Identity matrix '''

I = [1]

for i in range(M.shape[0]-1):
  I.append(0)

for i in range(M.shape[0]-1):
  tmp = [0]
  for j in range(M.shape[0]-1):
    if (i==j):
      tmp.append(1)
    else:
      tmp.append(0)
  I = np.vstack([I,tmp])




''' Determinant'''

def determinant(M):
  if(M.shape[0]==1):
    D = M[0]
    return D

  else:
    D = 0
    for i in range(M.shape[0]):
      for k in range(1,M.shape[0]):
        tmp = []
        for j in range(M.shape[0]):
          if (j!=i):
            tmp.append(M[k][j])
        tmp = np.array(tmp)
        if (k==1):
          ret = tmp
        else:
          ret = np.vstack([ret, tmp]) 
  
      if (i%2 == 0):
        D += M[0][i] * determinant(ret)
      else:
        D -= M[0][i] * determinant(ret)

    return D




''' Inverse matrix '''

def Inverse(M):
  M_inv = []
  for row in range(M.shape[0]):
    M_inv.append([0]*M.shape[0])

  for i in range(M.shape[0]):
    A = 0
    if (i==0):
      for j in range(M.shape[0]):
        
        for k in range(1,M.shape[0]):
          tmp = []
          for l in range(M.shape[0]):
            if (l!=j):
              tmp.append(M[k][l])
          tmp = np.array(tmp)
        
          if (k==1):
            A = tmp
          else:
            A = np.vstack([A,tmp]) 
  
        M_inv[i][j] = (-1)**(i+j) * determinant(A)

    else:
      for j in range(M.shape[0]):        
        
        for k in range(0,i):
          tmp = []
          for l in range(M.shape[0]):
            if (l!=j):
              tmp.append(M[k][l])
      
          tmp = np.array(tmp)
          if (k==0):
            A = tmp
          else:
            A = np.vstack([A,tmp])
    
        for k in range(i+1, M.shape[0]):
          tmp = []
          for l in range(M.shape[0]):
            if (l!=j):
              tmp.append(M[k][l])
      
          tmp = np.array(tmp)
          A = np.vstack([A,tmp])
    
        M_inv[i][j] = (-1)**(i+j) * determinant(A)
  
  M_inv = np.array(M_inv)
  return np.transpose(M_inv) / determinant(M)

M_modified = M + I * a

print(M)
print()
print(determinant(M))
print()
print(Inverse(M))
print()
print(M_modified)
print()
print(determinant(M_modified))
print()
print(Inverse(M_modified))
```
